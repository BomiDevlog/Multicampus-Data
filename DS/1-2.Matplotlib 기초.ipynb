# Matplotlib 기초

## 기본 구조

1. Figure : 그래프 전체 그림이라고 이해하시면 됩니다. 가장 큰 개념입니다.

2. Axes : 그래프가 그려지는 좌표 평면으로 이해하시면 됩니다.

             Figure 의 subplot 이 Axes 입니다. 한개의 Figure 에 여러개의 Axes 가 올 수 있습니다.

3. X axis : X 축 입니다.

4. Y axis : Y 축 입니다.

5. Tick : 눈금 입니다. Major 눈금과 Minor 눈금이 있습니다.

6. Spines : 그래프 테두리로 이해하시면 됩니다.

7. Line : 선 그래프(line plot)의 선입니다.

8. Markers : 산점도(scatter)에서 점 입니다.

9. Grid : 그래프 격자 입니다.

10. Title : 그래프 제목 입니다.

11. Label : 각 축이나 눈금 등에 붙이는 Text를 뜻합니다.

12. Legend : 범례라고 불리는 것인데, 각 그래프의 이름(label)을 알기 쉽게 별도로 표시해 놓는 것입니다.

```python
import matplotlib.pyplot as plt
# line plot 
# plot([x축데이터], [y축데이터])
plt.plot([1,2,3,4], [12, 43, 25, 15])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40])  # y값만 지정해주어도 그려줌 # x축은 자동으로 0,1,2 순. y축 값으로 적용
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('plotting') # 제목 지정
plt.plot([10, 20, 30, 40])
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('legend')
plt.plot([10, 20, 30, 40], label ='asc') # 증가를 의미하는 asc 범례 # label 해당 plot의 이름 지정
plt.plot([40, 30, 20, 10], label ='desc') # 감소를 의미하는 desc 범례
plt.legend() # 범례 표시 (우측에 각 라인의 설명란 표시)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('color') # 제목 설정
plt.plot([10, 20, 30, 40], color ='skyblue', label ='skyblue') # 하늘색 그래프
plt.plot([40, 30, 20, 10], 'pink', label ='pink') # 분홍색 그래프
plt.legend() # 범례 표시
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('color') # 제목 설정
plt.plot([10, 20, 30, 40], color ='slateblue', label ='slateblue') 
plt.plot([40, 30, 20, 10], 'mediumaquamarine', label ='mediumaquamarine') 
plt.legend() # 범례 표시
plt.show()
```

```python
# 헥사 코드 활용 https://www.pinterest.co.kr/pin/413275703309243526/ 
import matplotlib.pyplot as plt
plt.title('color') # 제목 설정
plt.plot([10, 20, 30, 40], color ='#f660ab', label ='hotpink') 
plt.plot([40, 30, 20, 10], '#360eff', label ='royalblue') 
plt.legend() # 범례 표시
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('linestyle') #제목 설정

plt.plot([10, 20, 30, 40], color ='r', linestyle ='--', label='dashed') # 빨간색 dashed 그래프
plt.plot([40, 30, 20, 10], color ='g', ls =':', label ='dotted') # 초록색 dotted 그래프
plt.legend() # 범례 표시
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('marker') # 제목 설정

plt.plot([10, 20, 30, 40], 'r.', label ='circle') # 빨간색 원형 마커 그래프
plt.plot([40, 30, 20, 10], 'g^', label ='triangle up')

# 초록색 삼각형 마커 그래프
plt.legend() # 범례 표시
plt.show()
```
